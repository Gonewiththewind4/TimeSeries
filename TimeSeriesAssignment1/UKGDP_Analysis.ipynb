loaddata about the UK GDP

```r
load(file="UKGDP.RData")
```

```r
GDPTimeSerises<-ts(UKGDP[,3],frequency = 4, start = c(1955,1))
plot(GDPTimeSerises,type = "l",col = "red",xlab = "t",ylab = "GDP",main = "Assignment1 -- GDP/t")
```

```r
library(forecast)
fitar1 <- Arima(GDPTimeSerises, order=c(0, 1, 1),include.constant = TRUE)
fitar2 <- Arima(GDPTimeSerises, order=c(1, 1, 0),include.constant = TRUE)
fitar3 <- Arima(GDPTimeSerises, order=c(1, 1, 1),include.constant =  TRUE)
```

```r
par(mfrow=c(3,1))
plot(fitar1$residuals, ylab="Residuals from ARIMA(0,1,1) fit")
plot(fitar2$residuals, ylab="Residuals from ARIMA(1,1,0) fit")
plot(fitar3$residuals, ylab="Residuals from ARIMA(1,1,1) fit")
```

```r
fcast_bj1 <- forecast(GDPTimeSerises, h=1)
fcast_bj2 <- forecast(GDPTimeSerises, h=2)
fcast_bj3 <- forecast(GDPTimeSerises, h=3)
fcast_bj4 <- forecast(GDPTimeSerises, h=4)
fcast_es <- ses(GDPTimeSerises, h=1)
fcast_holt <- holt(GDPTimeSerises, h=1)
```

```r
par(mfrow=c(1,2))
plot(fcast_holt)
plot(fcast_es)
```

```r
par(mfrow=c(2,2))
plot(fcast_bj1)
plot(fcast_bj2)
plot(fcast_bj3)
plot(fcast_bj4)
```
